```python
from sqlalchemy import create_engine, text
```

```python
database_name = 'prescription_drugs'    # Fill this in with your lahman database name

connection_string = f"postgresql://postgres:postgres@localhost:5432/{database_name}"
```

```python
engine = create_engine(connection_string)
```

```python
import pandas as pd
import numpy
```

```python
query = '''
SELECT *
FROM prescription
;
'''
with engine.connect() as connection:
    prescription = pd.read_sql(text(query), con = connection)

prescription
```

```python
query = '''
SELECT *
FROM drug
;
'''
with engine.connect() as connection:
    drug = pd.read_sql(text(query), con = connection)

drug
```

```python
query='''

select drug_name,
sum(case when opioid_drug_flag='Y' then total_drug_cost end) as opioid_cost,
sum(case when opioid_drug_flag='N' then total_drug_cost end)as non_opioid_cost 
from drug 
inner join prescription  using(drug_name)
group by drug_name;

'''

with engine.connect() as connection:
    drug_group = pd.read_sql(text(query), con = connection)
drug_group   
```

```python
total_opioid_cost=round(drug_group['opioid_cost'].sum(),2)
total_opioid_cost
```

```python
total_non_opioid_cost=round(drug_group['non_opioid_cost'].sum(),2)
total_non_opioid_cost
```

```python
ratio=round((total_opioid_cost / total_non_opioid_cost),3)
print('Ratio of Opioid and non opioidcosts:', ratio)
```

c. Are those who spend a higher ratio on opioids suffering from more deaths?

```python
query = '''
SELECT *
FROM overdose_deaths
;
'''
with engine.connect() as connection:
    overdose_deaths = pd.read_sql(text(query), con = connection)

overdose_deaths
```

```python
query='''
select sum(overdose_deaths) as deaths,fipscounty::int
from overdose_deaths
group by fipscounty;
'''
with engine.connect() as connection:
    overdose_deaths_county = pd.read_sql(text(query), con = connection)

overdose_deaths_county
```

```python
query='''
select sum(overdose_deaths) as deaths,year,fipscounty::int
from overdose_deaths
group by fipscounty,year;
'''
with engine.connect() as connection:
    overdose_deaths_county_year = pd.read_sql(text(query), con = connection)

overdose_deaths_county_year
```

```python
query='''

select drug_name,fipscounty::int, 
sum(case when opioid_drug_flag='Y' then total_drug_cost end) as opioid_cost,
sum(case when opioid_drug_flag='N' then total_drug_cost end)as non_opioid_cost 
from drug 
inner join prescription  using(drug_name)
inner join prescriber  using (npi)
inner join zip_fips on prescriber.nppes_provider_zip5=zip_fips.zip
group by drug_name,fipscounty;

'''

with engine.connect() as connection:
    drug_group_county = pd.read_sql(text(query), con = connection)
drug_group_county
```

```python
cost_county=drug_group_county.groupby(['fipscounty'])['opioid_cost'].sum()
cost_county
```

```python
total_deaths_county=pd.merge(overdose_deaths_county_year,cost_county,how ='inner',on=['fipscounty'])
```

```python
total_deaths_county
```

```python
total_deaths_county.sort_values(by='deaths',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(6, 6))
corr = total_deaths_county.corr(numeric_only=True)
cmap = sns.diverging_palette(220, 10, as_cmap=True)

mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

sns.heatmap(corr, cmap=cmap, mask = mask, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5});
```
